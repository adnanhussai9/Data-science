```python
df_sales=pd.read_csv("C:\Downloads\data fsds\sales_data_final.csv")
```

```python
df_sales
```

```python
df_sales["len_customer_name"]=df_sales["customer_name"].apply(len)
```

```python
df_sales
```

```python
def quantity_flag(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:
        return 'medium'
    else:
        return "high"
```

```python
df_sales["flag_quantity"]=df_sales["quantity"].apply(quantity_flag)
```

```python
df_sales.head()
```

```python
df_sales["square_qty"]=df_sales["quantity"].apply(lambda a:a**2)
```
