## India Meterological Department Rainfall Data(1901-2021)
https://www.imdpune.gov.in/Clim_Pred_LRF_New/Reports.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Rainfall_Data.xlsx")
df.head()
```

```python
avg_rain=[df.JUN.mean(),df.JUL.mean(),df.AUG.mean(),df.SEP.mean()]
explode=[0,0.1,0,0]
colors = sns.color_palette('bright')

plt.figure(figsize=(7,7))
plt.title(label="Rainfall According to Month",loc='left',fontstyle='italic')
plt.pie(avg_rain,labels=df.columns[1:-1],colors=colors,autopct="%.1f%%",shadow=True,explode=explode)
plt.show()
```

### On Average `July` recieves maximum amount of rainfall.

```python
plt.figure(figsize=(15,5))
plt.plot(df["YEAR"],df["JUN-SEP"],marker='o')
rain_mean=df["JUN-SEP"].mean()
plt.axhline(rain_mean,color="g",label=f"Mean Rainfall:{rain_mean:.2f}mm")
plt.xlabel("Year")
plt.ylabel("Total Rainfall")
plt.legend()
plt.show()
```
